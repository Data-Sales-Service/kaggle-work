```python
sorted_df = df.sort_values(["Total_Devices"], ascending=False)
```

# Top 10 Companies for most Mobile phones model

```python
plt.figure(figsize=(20,10))
colors = sns.color_palette('bright')[0:5]
plt.pie(sorted_df.head(20)['Total_Devices'], labels = sorted_df.head(20)['Company'], colors = colors, autopct='%.0f%%')
plt.show()
```

## Over share of mobile phones models

```python
plt.figure(figsize=(30,15))
colors = sns.color_palette('bright')[0:5]
plt.pie(sorted_df['Total_Devices'], labels = sorted_df['Company'], colors = colors, autopct='%.0f%%')
plt.show()
```

```python
plt.figure(figsize=(15,8))
ax = sns.barplot(x="Company",y="Total_Devices", data=sorted_df.head(50))
plt.xticks(rotation = 'vertical')
```

```python
plt.figure(figsize=(15,8))
ax = sns.barplot(x="Company",y="Total_Devices", data=sorted_df.tail(80))
plt.xticks(rotation = 'vertical')
```
