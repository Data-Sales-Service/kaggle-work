Importing Numpy, Pandas, and MatPlotLib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Reading Data EnglandWeather.csv

```python
data = pd.read_csv("/kaggle/input/englandweather/EnglandWeather.csv")
data
```

Converting the CSV file to Pandas DataFrame (Even though this step was not required).

```python
df = pd.DataFrame(data)
df
```

Describing the weather data.

```python
df.describe()
```

```python
df[["Formatted Date","Humidity"]]
```

Getting the date only from the complete formated date. It means that the timezone, and time in form of hours, minutes, seconds and miliseconds are truncated.
**This is done by spling the str by space**

After that formated date will only contains that date.

```python
df["Formatted Date"] = df["Formatted Date"].str.split(' ', expand = True)[0]
df
```

Dropping the null entries from the Dataframe, and making a copy in variable **df1**.

```python
df1 = df.dropna()
df1
```

Sort the values by **Humidity**, and creating a copy of **df1** to **df2**.

```python
df2 = df1.sort_values(by=["Humidity"])
df2[["Summary","Humidity"]].tail(20)
```

```python
df2["Summary"].unique()
```

```python
df2.groupby(["Summary"]).mean()
```

```python
df2.groupby(["Summary"]).mean()["Humidity"]
```

Creating a plot:
1. Creating a figure of size 15x10, with 80 dpi
2. Creating a Scatter graph between **Summary** and **Humidity**, and setting the color to **red**, with alpha as 0.5
3. Setting **xlabel** to **Summary**, **ylabel** to **Humidity**, title as **Humidity By Summary**. The xlabel, and ylabel are size of 20px, while the title has 25px size.
4. Set xticks to 15px, and rotate it to **90 degree**, as the labels are lengthy, and set the ylabel to 15 pixal.
5. Make the plot as **Grid**, and display graph with **show**.


```python
plt.figure(figsize = (15,10), dpi = 80)
plt.scatter(df2["Summary"], df2["Humidity"], color = ["red"],alpha = 0.5 )
plt.xlabel("\nSummary", fontsize = 20)
plt.ylabel("\nHumidity\n", fontsize = 20)
plt.title("Humidity By Summary\n", fontsize = 25)
plt.xticks(fontsize = 15 , rotation = 90)
plt.yticks(fontsize = 15)
plt.grid()
plt.show
```
