# 3. Rounding off `Time`

```python
train.head()
```

```python
train['time'] = pd.to_datetime(train['time']) # convert character into datetime object
```

```python
train_dt = train.set_index('time') #DateTimeIndex
```

```python
train_dt.index
```

```python
train_dt.index.round('S')
```
