# EDA of Ted Talks

## Setup
### Importing Liberaries

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import matplotlib.pyplot as plt
```

### Loading data

```python
df = pd.read_csv('/kaggle/input/ted-talks/data.csv')
```

### Displaying first five records

```python
df.head()
```

### Displaying dinmentions of the data

```python
df.shape
```

## Most Viewed Ted Talks

```python
ten_most_viewed_talks = df.sort_values("views").tail(10)
```

```python
ten_most_viewed_talks[['author','views']].head(10)
```

```python
sns.set_theme()
plt.figure(figsize=(15,8))
ax = sns.barplot(x="author", y="views", data=ten_most_viewed_talks[['author','views']].head(10))
ax.tick_params(axis='x', rotation=45)
```

## Most Liked Ted Talks

```python
ten_most_liked_talks = df.sort_values("likes").tail(10)
```

```python
ten_most_liked_talks[['author','likes']].head(10)
```

```python
sns.set_theme()
plt.figure(figsize=(15,8))
ax = sns.barplot(x="author", y="likes", data=ten_most_liked_talks[['author','likes']].head(10))
ax.tick_params(axis='x', rotation=45)
```
