# Striping a string in Python

There are three methods to strip a string in Python, means to remove spaces around the string, or any charactors.
1. stip
2. rstrip
3. lstrip

The **strip()**, **lstrip()**, and **rstrip()** methods in Python are used to remove specified characters from the beginning, end, or both ends of a string.

* **strip([chars]):** Removes the characters specified in **chars** from the beginning and end of the string. If chars is not provided, it removes all whitespaces (spaces, tabs, and newlines) from the beginning and end of the string.
* **lstrip([chars]):** Removes the characters specified in **chars** from the beginning of the string. If chars is not provided, it removes all whitespaces from the beginning of the string.
* **rstrip([chars]):** Removes the characters specified in **chars** from the end of the string. If chars is not provided, it removes all whitespaces from the end of the string.

```python
text = "  \t   Hello World   \n  "
print(text.strip())    # Output: "Hello World"
print(text.lstrip())   # Output: "Hello World   \n  "
print(text.rstrip())   # Output: "  \t   Hello World"
```

## Examples

```python
# 1. Removing specific characters from the left side of a string using lstrip():
string = "  ,,,,,,hello world,"
new_string = string.lstrip(", ")
print(new_string) # Output: "hello world,"
```

```python
# 2. Removing specific characters from the right side of a string using rstrip():
string = "hello world,,,,,,   "
new_string = string.rstrip(", ")
print(new_string) # Output: "hello world"
```

```python
# 3. Removing specific characters from both sides of a string using strip():
string = "  ,,,,,,hello world,,,,,,   "
new_string = string.strip(", ")
print(new_string) # Output: "hello world"
```

```python
# 4. Removing leading or trailing whitespace using strip(), lstrip(), or rstrip():
string = "   hello world   "
new_string = string.strip()
print(new_string) # Output: "hello world"
```

```python
# 5. Stripping a string of all digits:
string = "123hello456world789"
new_string = string.strip("0123456789")
print(new_string) # Output: "hello456world"
```

```python
# 6. Strip leading zeros from a string:
s = '00001234'
s = s.lstrip('0')
print(s)  # Output: '1234'
```

```python
# 7. Remove a prefix and/or suffix from a string:
s = 'foobar'
s = s.lstrip('foo')
print(s)  # Output: 'bar'

s = 'foobar'
s = s.rstrip('bar')
print(s)  # Output: 'foo'
```

```python
# 8. Remove all punctuation from the beginning and end of a string:
import string

s = "???Hello, World!???"

# Remove punctuation from the beginning of the string
s = s.lstrip(string.punctuation)

# Remove punctuation from the end of the string
s = s.rstrip(string.punctuation)

print(s)  # Output: Hello, World
```

```python
# 9. Removing prefixes or suffixes from a string: You can use lstrip() or rstrip() to remove a specific prefix or suffix from a string, respectively. For example:

string = "### Hello world ###"
stripped_string = string.lstrip("# ")
print(stripped_string)  # Output: "Hello world ###"

stripped_string = string.rstrip("# ")
print(stripped_string)  # Output: "### Hello world"

stripped_string = string.strip("# ")
print(stripped_string)  # Output: "Hello world"
```

```python
# 10. Removing specific characters from the beginning or end of a string: You can use lstrip() or rstrip() to remove specific characters from the beginning or end of a string, respectively. For example:
string = ".....Hello world....."
stripped_string = string.lstrip(".")
print(stripped_string)  # Output: "Hello world....."

stripped_string = string.rstrip(".")
print(stripped_string)  # Output: ".....Hello world"

stripped_string = string.strip(".")
print(stripped_string)  # Output: "Hello world"
```

```python
# 11. Removing multiple whitespace characters from a string: You can use strip() to remove multiple whitespace characters from a string, such as tabs, newlines, and spaces. For example:
string = "   Hello \t world\n\n  "
stripped_string = string.strip()
print(stripped_string)  # Output: "Hello    world"
```
