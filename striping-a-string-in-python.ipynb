# Striping a string in Python

There are three methods to strip a string in Python, means to remove spaces around the string, or any charactors.
1. stip
2. rstrip
3. lstrip

The **strip()**, **lstrip()**, and **rstrip()** methods in Python are used to remove specified characters from the beginning, end, or both ends of a string.

* **strip([chars]):** Removes the characters specified in **chars** from the beginning and end of the string. If chars is not provided, it removes all whitespaces (spaces, tabs, and newlines) from the beginning and end of the string.
* **lstrip([chars]):** Removes the characters specified in **chars** from the beginning of the string. If chars is not provided, it removes all whitespaces from the beginning of the string.
* **rstrip([chars]):** Removes the characters specified in **chars** from the end of the string. If chars is not provided, it removes all whitespaces from the end of the string.

```python
text = "  \t   Hello World   \n  "
print(text.strip())    # Output: "Hello World"
print(text.lstrip())   # Output: "Hello World   \n  "
print(text.rstrip())   # Output: "  \t   Hello World"
```
