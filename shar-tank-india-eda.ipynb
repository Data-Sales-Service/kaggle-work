```python
def synergy(data):
  list = ['anupam_deal','aman_deal','namita_deal','vineeta_deal','peyush_deal','ghazal_deal','ashneer_deal']
  for i in list:
    deal = data[['amount_per_shark','equity_per_shark']][data[i]==1]
    print("\n{} deals with {}".format(len(deal),i[:-5]))
    print(deal)
```

```python
ash_grover = df[df['ashneer_deal']==1]
ash_grover
```

```python
ash_grover['amount_per_shark'].sum()
```

```python
synergy(ash_grover)
```

```python
am_gupta = df[df['aman_deal']==1]
```

```python
synergy(am_gupta)
```

```python
eqt = df.groupby('ashneer_deal')['equity_per_shark'].sum().loc[1]
amt = df.groupby('ashneer_deal')['amount_per_shark'].sum().loc[1]
print("Total equity buy on shark tank",eqt,'%')
print("Total amount invested on shark tank",amt,"lakhs")
```
